```python
import os
import glob

import numpy as np
import pandas as pd
from functools import reduce

from pymfe.mfe import MFE


def read_mtf_datasets(paths, how="mtf"):
    def read(path, idx):
        aux = pd.read_pickle(path)
        aux = pd.DataFrame([aux[idx]], columns=aux[0])
        aux["dataset"] = get_dataset_name(path)
        return aux

    if how == "mfe":
        n=1
    elif how == "time":
        n=2
    else:
        n=0

    return pd.concat([read(i, n) for i in paths])
```

## Meta-features

```python
def get_dataset_name(path):
    result = path.split("/")[-1].split("_")
    return result[0] + "_" + result[1]

mf_runned_path = glob.glob("../../../results/metafeature_extraction/*/1/*.pkl")
mf_path = np.array(mf_runned_path)
df_mtf = read_mtf_datasets(mf_path, how="mfe")
df_mtf
```

```python
df_mtf.to_csv("meta_feature_dataset.csv", index=False)
```

## Meta features scoring time

```python
df_mtf_time = read_mtf_datasets(mf_path, how="time")
df_mtf_time
```

```python
df_mtf_time.to_csv("meta_feature_time_spent.csv", index=False)
```

## Meta feature information

```python

cols = df_mtf.columns.tolist()
cols.remove("dataset")

group_map = reduce(
    lambda x, y: x | y,
    [{k:i for k in MFE.valid_metafeatures(groups=i)} for i in  sorted(MFE.valid_groups())[::-1]]
)
col_summary = [i.split(".")[1] if len(i.split(".")) > 1 else None for i in cols]
col_name = [i.split(".")[0] for i in cols]
col_group = [group_map.get(i, None) for i in col_name]
df_meta_feaures_info = pd.DataFrame({
    "meta-feature": col_name,
    "summary": col_summary,
    "group": col_group,
    "full-name": cols,
 })

df_meta_feaures_info
```
