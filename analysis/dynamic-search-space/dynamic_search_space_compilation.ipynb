## Compilation of Results

```python
RS = pd.read_csv("experiments/rs_automl.csv")


def best_pipe_test_performance(method, dataset, time, seed):
  aux = method.query(f"seed == {seed} and dataset == '{dataset}' and cumulative_sum_cv_duration_mtf <= {time}")
  if len(aux) != 0:
    return aux.sort_values(by="cv_median_f1_weighted", ascending=False).head(1)["f1_weighted_test"].values[0]
  return 0.0

def statistical_test_scv(df, time):
  aux_ = df\
    .query(f"time == {time}")\
    .pivot(index=["dataset", "seed"],columns="method", values="performance")

  aux_.to_csv(PATH + f"meta-automl-all-{time}.csv", index=False)

  return aux_
```

```python
def performance_over_time(experiment_name, experiment_data, step=60, end=36000):
  step = 60
  methods = {
      experiment_name: experiment_data,
  }
  seeds = RS["seed"].unique().tolist()
  datasets = RS["dataset"].unique().tolist()

  r = []
  for t in range(step, end, step):
    for m_name, m in methods.items():
      for s in seeds:
        for d in datasets:
          p = best_pipe_test_performance(m, d, t, s)
          r.append({
              "method": m_name,
              "dataset": d,
              "time": t,
              "seed": s,
              "performance": p
          })
  aux_ = pd.DataFrame(r)
  aux_.to_csv(PATH + f"{experiment_name}_over_time_{step}_{end}.csv", index=False)
  return aux_
```

```python
meta_automl_99 = pd.read_csv("experiments/meta_automl_omega_0.99.csv")
performance_over_time("RS-mtl-99", meta_automl_99, step=60, end=36060)
```
