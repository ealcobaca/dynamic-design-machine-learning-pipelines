```python
df_datasets = pd.read_csv("../meta-feature-analysis/datasets_meta_data.csv", index_col=0)\
    .rename(columns={
        "Number of Examples": "number of examples",
        "Number of Features": "number of features",
        "Number of class": "number of classes",
        "Minority Class %": "minority class (%)"
    })
df_datasets
```

```python
df_datasets["dataset"] = ["dataset_"+str(i) for i in df_datasets["OpenML ID"]]
# select all
df_datasets = df_datasets[df_datasets["dataset"].isin(D)]
# tag train/test
df_datasets["group"] = ["train" if i in D_train else "test"  for i in df_datasets["dataset"].to_list()]
```

```python
df_datasets["group"].value_counts()
```

```python
def save_pdf(name, g, extention=".pdf", path="./"):
  g.savefig(path+name+extention)
```

```python
columns = [
    "number of examples (log10)",
    "number of features (log10)",
    "number of classes (log10)",
    "minority class (%)"
]
df_datasets["number of examples (log10)"] = np.log10(df_datasets["number of examples"])
df_datasets["number of features (log10)"] = np.log10(df_datasets["number of features"])
df_datasets["number of classes (log10)"] = np.log10(df_datasets["number of classes"])


g = sns.PairGrid(df_datasets[columns])
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)

save_pdf("datasets_pair_grid_all_fill", g, extention=".pdf", path=PATH)
```

```python
g = sns.PairGrid(df_datasets[columns])
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=False)
g.map_diag(sns.histplot, kde=True)
save_pdf("datasets_pair_grid_all_notfill", g, extention=".pdf", path=PATH)
```

```python
for i in columns:
  g=sns.displot(df_datasets[[i, "group"]], x=i, hue="group", kind="kde", fill=True, hue_order=["train", "test"])
  save_pdf(f"datasets_distplot_{i}", g, extention=".pdf", path=PATH)
```

```python
for i in columns:
  for j in columns:
    if i == j: continue
    sns.jointplot(
        data=df_datasets,
        x=i, y=j,
        hue="group",
        kind="kde",
        hue_order=["train", "test"]
    )
    save_pdf(f"datasets_jointplot_{i}_vs_{j}", g, extention=".pdf", path=PATH)
```

```python
g = sns.PairGrid(df_datasets[columns])
g.map_upper(sns.scatterplot, hue=df_datasets["group"], hue_order=["train", "test"], style=df_datasets["group"], markers=["o","X"])
g.map_lower(sns.kdeplot, fill=False, hue=df_datasets["group"], hue_order=["train", "test"])
g.map_diag(lambda x, **kwargs: sns.histplot(x=x, hue_order=["train", "test"], hue=df_datasets["group"], kde=True, legend=False, multiple="stack"))

handles = g._legend_data.values()
labels = g._legend_data.keys()
g.fig.legend(handles=handles, labels=labels, loc='lower center', ncol=2)
g.fig.subplots_adjust(bottom=0.1)

save_pdf("datasets_pair_grid_group_notfill", g, extention=".pdf", path=PATH)
```
