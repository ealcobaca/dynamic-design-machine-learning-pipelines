## Meta Target Definition

```python
def cross_validation_selection_by_median(meta_knowledge, measure):
  """
    Selec model via cross-validation based on best median.
  """

  median = meta_knowledge\
    .groupby(["seed", "config_id", "dataset"])[measure]\
    .median()\
    .to_frame("median_perf")\
    .reset_index()
  time = meta_knowledge\
    .groupby(["seed", "config_id", "dataset"])["duration"]\
    .sum()\
    .to_frame("sum_cv_duration")\
    .reset_index()
  meta_knowledge_ = meta_knowledge\
    .merge(median, how="inner", on=["seed", "config_id", "dataset"])
  meta_knowledge_ = meta_knowledge_\
    .merge(time, how="inner", on=["seed", "config_id", "dataset"])
  meta_knowledge_["diff_performance"] = np.abs(meta_knowledge_["median_perf"].fillna(0.0) - meta_knowledge_[measure].fillna(0.0))
  meta_knowledge_["first"] = meta_knowledge_\
    .groupby(["seed", "config_id", "dataset"])["diff_performance"]\
    .rank(method="first")
  meta_knowledge_ = meta_knowledge_[meta_knowledge_["first"] == 1]

  pipelines = meta_knowledge_["classifier:__choice__"]+" + "+meta_knowledge_["feature_preprocessor:__choice__"]
  meta_knowledge_ = meta_knowledge_ \
    .assign(pipelines=pipelines) \
    .rename(columns={"classifier:__choice__": "classifier", "feature_preprocessor:__choice__": "preprocessor"})

  return meta_knowledge_.drop(columns=["start_time", "end_time", "budget"]) #remove because can not make sense after aggregation
```

```python
meta_knowledge = cross_validation_selection_by_median(meta_knowledge=meta_knowledge_, measure="f1_weighted_val")
meta_knowledge
```

```python
meta_knowledge.to_csv("meta_knowledge_cv_f1_weighted.csv", index=False)
```

```python
def meta_target(meta_knowledge, metric, target):
  if not isinstance(target, list):
    target = [target]

  metric_val = metric + "_val"
  df_ = cross_validation_selection_by_median(meta_knowledge=meta_knowledge, measure=metric_val)
  df_["rank"] = df_\
      .groupby(["dataset"] + target)[metric_val]\
      .rank(method="first", ascending=False, na_option="bottom")
  return df_.query("rank == 1")[["dataset"] + target + [metric+"_val", metric+"_test", metric+"_train"]]
```

```python
meta_target_pipe = meta_target(
    meta_knowledge=meta_knowledge_,
    metric="f1_weighted",
    target=["preprocessor", "classifier"]
)
```

## Meta-dataset

```python
def generate_meta_dataset(D_train, d_test, meta_knowledge, meta_target, meta_feature):
  minn = -1e10
  maxx = 1e10

  # fill pipeline without performance with 0
  # fill nan meta-feature with 0 and replace +-inf
  meta_dataset_ = meta_target.fillna(0)\
    .merge(
        meta_feature\
          .replace([np.inf], maxx)\
          .replace([-np.inf], minn),
        how="inner", on="dataset")\
    .fillna(0.0)

  aux_cl, aux_pr, aux_p = pipeline_statistics_meta_features(
        meta_knowledge, exclusion_list=d_test)

  # fill nan performance with 0
  meta_dataset_ = meta_dataset_\
    .merge(aux_cl, on="classifier", how="left")\
    .merge(aux_pr, on="preprocessor", how="left")\
    .merge(aux_p, on=["classifier", "preprocessor"], how="left")\
    .fillna(0.0)

  meta_dataset_["classifier-num"] = meta_dataset_["classifier"].astype("category").cat.codes
  meta_dataset_["preprocessor-num"] = meta_dataset_["preprocessor"].astype("category").cat.codes

  train_idx = meta_dataset_["dataset"].isin(D_train)
  test_idx = meta_dataset_["dataset"].isin(d_test)

  return meta_dataset_[train_idx], meta_dataset_[test_idx]
```

### Pipeline Statistics Meta Feature

```python
def pipeline_statistics_meta_features(meta_knowledge, exclusion_list=None):
  """ Use exclusion_list with list of dataset to remove."""

  if exclusion_list:
    idx = ~meta_knowledge["dataset"].isin(exclusion_list)
    meta_knowledge = meta_knowledge[idx]

  aux_cl = meta_knowledge \
      .groupby(["classifier"]) \
      .agg(
          cl_min_perf=("median_perf", "min"),
          cl_max_perf=("median_perf", "max"),
          cl_median_perf=("median_perf", "median"),
          cl_mean_perf=("median_perf", "mean"),
          cl_std_perf=("median_perf", "std")
      )

  aux_pr =  meta_knowledge \
      .groupby(["preprocessor"]) \
      .agg(
          pr_min_perf=("median_perf", "min"),
          pr_max_perf=("median_perf", "max"),
          pr_median_perf=("median_perf", "median"),
          pr_mean_perf=("median_perf", "mean"),
          pr_std_perf=("median_perf", "std")
      )

  aux_p =  meta_knowledge \
      .groupby(["classifier", "preprocessor"]) \
      .agg(
          p_min_perf=("median_perf", "min"),
          p_max_perf=("median_perf", "max"),
          p_median_perf=("median_perf", "median"),
          p_mean_perf=("median_perf", "mean"),
          p_std_perf=("median_perf", "std")
      )

  return aux_cl, aux_pr, aux_p
```

```python
# split dataset set
D_train_meta, d_test_meta = generate_meta_dataset(D_train, d_test, meta_knowledge, meta_target, meta_feature)

# reapeat experiments with different seeds
result_meta_learners={}
for seed in SEEDS:
  models = {
    "knn": Pipeline([('scaler', MinMaxScaler()), ('knn', KNeighborsRegressor())]),
    "svm": Pipeline([('scaler', MinMaxScaler()), ('svm', SVR())]),
    "mlp": Pipeline([('scaler', MinMaxScaler()), ('mlp', MLPRegressor(random_state=seed))]),
    "linear": Pipeline([('scaler', MinMaxScaler()), ('linear', Lasso(random_state=seed))]),
    "rf": Pipeline([('rf', RandomForestRegressor(random_state=seed))]),
    "baseline": Pipeline([('dummy', DummyRegressor(strategy="mean"))])
  }

  result = {}
  for k, m in models.items():
    print()
    X_train = D_train_meta[METAFEATURES_ALL]
    y_train = D_train_meta[METACLASS]
    result[k] = cross_validate(
        m,
        X_train,
        y_train,
        cv=CV,
        n_jobs=10,
        scoring=MEASURES,
        return_train_score=True
    )
  result_meta_learners[seed] =  result


with open('result_meta_learners.pickle', 'wb') as handle:
    pickle.dump(result_meta_learners, handle)
```
